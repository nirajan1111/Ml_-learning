## matplotlib example workflow

```python
# 0. import Matplotlib 
%matplotlib inline 
import matplotlib.pyplot as plt

# 1. Prepare data 
x=[1,2,3,4]
y=[11,22,33,44]

# 2. setup plot
fig,ax= plt.subplots(figsize=(5,5))

# 3. plot data 
ax.plot(x,y)

#4. customize plot
ax.set(title="A plot ",
       xlabel="x-axis",
       ylabel="y-axis" )

#5. Save and show 
fig.savefig("images/a-simple-plot.png")
```

## Two options for subplots

```python
# Subplot Option 1
fig,((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# Plot to each different axis
ax1.plot(x,x/4)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_buter_prices.keys(),nut_buter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots Option 2
fig, ax= plt.subplots(nrows=2,ncols=2, figsize=(10,5))

# Plot to different indexes
ax[0,0].plot(x,x/4)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_buter_prices.keys(),nut_buter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use (pyplot vs matplotlib oo method)

* When plotting quickly , okay to use pyplot method 
* When plotting something advanced , use the OO method

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter", x="age", y="chol", c="target",figsize=(10,6))
```

```python
# OO method mixed pyplot
fix, ax =plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax)
#ax.set_xlim([45,100]) # give the xlimit 
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,8))

# Plot the data
scatter = ax.scatter(x= over_50["age"], y= over_50["chol"],c=over_50["target"]);


# customize the plot
ax.set(title= "heart disease",
      xlabel="Age ",
      ylabel="Cholestrol"
      );

# adding a legend
ax.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal line 
ax.axhline(over_50["chol"].mean(),
          linestyle='-'
          )
```

```python
over_50
```

```python
# Subplots  of chol, age ,thalach
fig, (ax0, ax1)= plt.subplots(nrows =2, ncols=1, figsize=(10,10))

# Add data to ax0
scatter= ax0.scatter(x=over_50["age"], y= over_50["chol"], c= over_50["target"])

# Customize ax0
ax0.set(title= "HEart diseas adn cholestro. level" , xlabel= "age", ylabel="cholestrol");

# Add a legend 
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(y = over_50["chol"].mean(),
          linestyle='--'
          );

# Add data to ax1
scatter= ax1.scatter(x=over_50["age"], y= over_50["thalach"], c= over_50["target"])

# Customize ax1
ax1.set(title= "HEart diseas adn thalach. level" , xlabel= "age", ylabel="thalach");

# Add a legend 
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax1.axhline(y = over_50["thalach"].mean(),
          linestyle='--'
          );


#Add a title to the fig
fig.suptitle("Heart disease analysis, ", fontsize =18, fontweight="bold");
```

```python
plt.style.use( 'seaborn-v0_8-whitegrid',)
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x= over_50["age"], 
                     y= over_50["chol"],
                     c=over_50["target"],
                    cmap= "winter");# this changes the colour scheme


# customize the plot
ax.set(title= "heart disease",
      xlabel="Age ",
      ylabel="Cholestrol"
      );

# adding a legend
ax.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal line 
ax.axhline(over_50["chol"].mean(),
          linestyle='-'
          )
```

```python
# Subplots  of chol, age ,thalach
fig, (ax0, ax1)= plt.subplots(nrows =2, ncols=1, figsize=(10,10))

# Add data to ax0
scatter= ax0.scatter(x=over_50["age"], y= over_50["chol"], c= over_50["target"], cmap= "winter")

# Customize ax0
ax0.set(title= "HEart diseas adn cholestro. level" , xlabel= "age", ylabel="cholestrol" );
ax0.set_xlim([50,80])
# Add a legend 
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(y = over_50["chol"].mean(),
          linestyle='--'
          );

# Add data to ax1
scatter= ax1.scatter(x=over_50["age"], y= over_50["thalach"], c= over_50["target"], cmap= "summer")

# Customize ax1
ax1.set(title= "HEart diseas adn thalach. level" , xlabel= "age", ylabel="thalach");

# Add a legend 
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax1.axhline(y = over_50["thalach"].mean(),
          linestyle='--'
          );


#Add a title to the fig
fig.suptitle("Heart disease analysis, ", fontsize =18, fontweight="bold");
```
